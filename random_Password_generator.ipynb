```python
import string
import random

# print(' Enter the number of passwords you want :')
numOfPass = int(input(' Enter the number of passwords you want '))

# print(" Enter the length of digits you want as Password : ")
digits = int(input(" Enter the length of digits you want as Password : "))

for num in range(0, numOfPass):

    lower = string.ascii_lowercase

    upper = string.ascii_uppercase

    symbol = string.punctuation

    number = string.digits

    password = lower + upper + number + symbol

    randPassword = random.sample(password, digits)
    
# printing randPassword will give you the passwords but in the form of a list, that you don't want, so use .join() method.

    finalPassword = ''.join(randPassword)
    
    print(finalPassword)

```
